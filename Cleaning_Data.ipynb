# Missing Data Handling

```python
data['reviews_per_month'].fillna(0, inplace=True)

```

```python
data['last_review'] = pd.to_datetime(data['last_review'], format='%d-%m-%Y', errors='coerce')
data['last_review'].fillna(pd.Timestamp('1900-01-01'), inplace=True)

```

```python
# Replace nulls in 'name' column
data['name'].fillna('Unknown', inplace=True)

# Replace nulls in 'host_name' column
data['host_name'].fillna('Unknown', inplace=True)
```

# Standardization



```python
data["name"] = data["name"].str.lower()
data["host_name"] = data["host_name"].str.lower()
data["neighbourhood_group"] = data["neighbourhood_group"].str.lower()
data["neighbourhood"] = data["neighbourhood"].str.lower()
data["room_type"] = data["room_type"].str.lower()
```

# Outlier Detection:

```python
num_cols = ['price', 'minimum_nights', 'number_of_reviews',
            'reviews_per_month', 'calculated_host_listings_count',
            'availability_365']

print("=== Outlier Detection Using IQR Method ===\n")

for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Find outliers
    outliers = data[(data[col] < lower) | (data[col] > upper)]

    print(f"Column: {col}")
    print(f"Outliers found: {outliers.shape[0]}")
    print("---------------------------")
```

# Outlier Removal:

```python
data_clean = data.copy()

for col in num_cols:
    Q1 = data_clean[col].quantile(0.25)
    Q3 = data_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_clean = data_clean[(data_clean[col] >= lower) & (data_clean[col] <= upper)]

data_clean.shape
```
